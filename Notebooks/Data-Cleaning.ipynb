```python
import numpy as np
import pandas as pd
from datetime import datetime
```

```python
# let't upload our data
```

```python
stores = pd.read_csv('../Data/Raw-data/stores.csv')
stores.head()
```

```python
features = pd.read_csv('../Data/Raw-data/features.csv', parse_dates=['Date'])
features.head()
```

```python
train = pd.read_csv('../Data/Raw-data/train.csv', parse_dates=['Date'])
train.head()
```

```python
# let's merge our data
```

```python
data = train.merge(features, on=['Store', 'Date'], how='inner').merge(stores, on=['Store'], how='inner')
data.head()
```

```python
# lets drop dublicated columns
```

```python
data.drop(['IsHoliday_x'], axis=1, inplace=True)
```

```python
data.rename(columns={'IsHoliday_y':'IsHoliday'},inplace=True)
```

Walmart gave markdown columns to see the effect if markdowns on sales. When I check columns, there are many NaN values for markdowns. I decided to change them with 0, because if there is markdown in the row, it is shown with numbres. So, if I can write 0, it shows there is no markdown at that date.

```python
data = data.fillna(0)
```
